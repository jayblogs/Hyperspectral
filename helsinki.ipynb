```python
import numpy as np
```

```python
array = raster.read()
stats = []
```

```python
for band in array:
    stats.append({
    'min': band.min(),
    'mean': band.mean(),
    'median': np.median(band),
    'max': band.max()})
```

```python
print(stats)
```
